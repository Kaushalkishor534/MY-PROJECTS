## 2.1 multi leagues
* https://www.football-data.co.uk/mmz4281/2526/E0.csv
* https://www.football-data.co.uk/mmz4281/2526/E1.csv
* https://www.football-data.co.uk/mmz4281/2526/E2.csv
* https://www.football-data.co.uk/mmz4281/2526/E3.csv
* https://www.football-data.co.uk/mmz4281/2526/EC.csv

```python
#  creating list of leagues in 2026
leagues =['E0','E1','E2','E3','EC']
Frames=[]
```

```python
# LOOP FOR READ MULTI LEAGUES and store in frame list
for league in leagues:
    df=pd.read_csv(root+'2526/'+league+'.csv',encoding='unicode_escape')
    Frames.append(df) ## stored in frames the df otherwise (eo can be replaced by e1 in df) 
```

```python
# length of frame
len(Frames) ## have 5 leagues csv data
```

```python
Frames[4]
```

# 2.2 for multiple seasons(2020,21,22...)

```python
for season in range(20,26):
    print(str(season)+str(season+1))
```

```python
#  creating list of leagues in 2026
leagues =['E0','E2','E3']
Frames=[]
# LOOP For range of seasons and insert new column
for league in leagues:
    for season in range (20,26):
        df=pd.read_csv(root + str(season)+str(season+1)+'/'+league+'.csv')
        df.insert(1,'season',season)
        Frames.append(df) ## stored in frames the df otherwise (eo can be replaced by e1 in df) 
```

```python
# total frames (6seasons * 3 leagues)
len(Frames)
```

```python
# show 1st and last element (new season column was created)
Frames[17]
```

# 3. organize all in dictionary

```python
# creating dictonary with name of league as a key
dict_countries={'spanish la liga':'SP1','spanish segunda division'
                :'SP2','german bundesliga1 ':'D1','german bundesliga2 ':'D2',
                'english premier league':'E0','english league 1':'E2','english league 2':'E3'
}
```

```python
# get a dictionary element
dict_countries['spanish la liga']
```

```python
for league in dict_countries:
    print (dict_countries[league])
```

```python
dict_historical_data={}
#leagues =['E0','E2','E3']

# LOOP For range of seasons and insert new column
for league in dict_countries:
    Frames=[]
    for season in range (20,26):
        df=pd.read_csv(root + str(season)+str(season+1)+'/'+dict_countries[league]+'.csv')
        df.insert(1,'season',season)
        Frames.append(df) 
    df_concat = pd.concat(Frames)
    dict_historical_data[league]=df_concat
```
